```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_random_values(n=20, min_val=0, max_val=100):
    return np.random.randint(min_val, max_val + 1, size=n).tolist()
```

```python
def calculate_deepest_lake(heights):
    max_depth = 0
    deepest_lake_range = []

    for i in range(len(heights) - 1):
        for j in range(i + 2, len(heights)):
            if heights[i] > heights[j]:
                min_height = min(heights[i:j+1])
                depth = min(heights[i], heights[j]) - min_height
                if depth > max_depth:
                    max_depth = depth
                    deepest_lake_range = (i, j)

    return max_depth, deepest_lake_range
```

```python
def visualize(heights, deepest_lake_range):
    plt.figure(figsize=(10, 5))
    x = np.arange(len(heights))

    plt.plot(x, heights, marker='o', linestyle='-', color='blue', label='Terrain')

    if deepest_lake_range:
        i, j = deepest_lake_range
        plt.plot(x[i:j+1], heights[i:j+1], marker='o', linestyle='-', color='red', label='Deepest Lake')
        plt.hlines([min(heights[i], heights[j])], i, j, colors='green', linestyles='dashed', label='Lake Level')

    plt.xlabel('Index')
    plt.ylabel('Height')
    plt.title('Visualization of Heights and Deepest Lake')
    plt.legend()
    plt.show()
```

```python
# Generate data
heights = generate_random_values()

# Calculate the deepest lake
deepest_lake_depth, lake_range = calculate_deepest_lake(heights)
print(f"Deepest Lake Depth: {deepest_lake_depth}")
print(f"Deepest Lake Range: {lake_range}")

# Visualize the data
visualize(heights, lake_range)
```
